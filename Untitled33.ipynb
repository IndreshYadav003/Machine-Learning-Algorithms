```python
#unsupervised Machine Learning Algorithms ! 1 K-Means Clustering
import seaborn as sns
from sklearn.datasets import make_blobs
x,y=make_blobs(n_samples=1000,n_features=2,centers=4,random_state=42)
sns.scatterplot(x=x[:,0],y=x[:,1])
```

```python

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
# k=4
'''Elbow method'''
wcss=[]
for k in range(1,21):
  kmeans=KMeans(n_clusters=k)
  kmeans.fit_predict(x)
  wcss.append(kmeans.inertia_)
sns.lineplot(x=range(1,21),y=wcss,markers='o')
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python

from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=k,random_state=42)
kmeans.fit(x)
y_pred=kmeans.predict(x)
sns.scatterplot(x=x[:,0],y=x[:,1],hue=y_pred)
```

```python

# iris data set operations

import seaborn as sns
from sklearn.datasets import load_iris
iris=load_iris()
x=iris.data
y=iris.target
sns.scatterplot(x=x[:,0],y=x[:,1])
```

```python
# scale the cluster
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
sns.scatterplot(x=x_scaled[:,0],y=x_scaled[:,1])
```

```python


from sklearn.cluster import KMeans
k=3
kmeans=KMeans(n_clusters=k,random_state=42)
kmeans.fit(x_scaled)
y_pred=kmeans.predict(x_scaled)
sns.scatterplot(x=x_scaled[:,0],y=x_scaled[:,1],c=y_pred)
```

```python
import numpy as np

X = np.array(data)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

print(labels)
print(kmeans.cluster_centers_)
```
